```python
def PageRank(graph, d,n, iter):
    node_list = list(graph.nodes)
    pagerank = dict.fromkeys(node_list, 1.0)
    print(node_list)
    for i in range(iter):
      for node in node_list:
          
          in_edges = graph.in_edges(node)
          in_neighbors = [i for (i, j) in in_edges]
          
          #node = str(node)
          pagerank_sum = sum((pagerank[in_n] / len(graph.out_edges(in_n))) for in_n in in_neighbors)
          
          random_walk = d / n
          
          pagerank[node] = random_walk + (1-d) * pagerank_sum
    return pagerank
```

# **HITS ALgorithm**

```python
def HITS(graph, iter):
    node_list = list(graph.nodes)
    auth = dict.fromkeys(node_list, 1.0)
    hub = dict.fromkeys(node_list, 1.0)
    for i in range(iter):
      for node in node_list:
        in_edges = graph.in_edges(node)
        in_neighbors = [i for (i, j) in in_edges]
        auth[node] = sum(hub[in_node] for in_node in in_neighbors)

        out_edges =graph.out_edges(node)
        out_neighbors = [i for (i, j) in out_edges]
        hub[node] = sum(auth[out_node] for out_node in out_neighbors)


        auth_sum = sum(auth[n] for n in node_list)
        hub_sum = sum(hub[n] for n in node_list)

        for node in node_list:
            auth[node] /= auth_sum
            hub[node] /= hub_sum
        
    return auth,hub
```

```python
auth,hub = HITS(G,50)
```
